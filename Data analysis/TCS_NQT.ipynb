```python
# Problem Statement:
# ------------------
# Write a Python program that takes a string S as input.
# The string consists only of two characters: '*' and '#'.
# The length of the string is variable.
#
# Your task is to determine the minimum number of characters
# ('*' or '#') that need to be added to make the string valid.
#
# A string is considered valid if the number of '*' and '#'
# characters in the string are equal.
#
# Output Rules:
# -------------
# 1. If the number of '*' is greater than '#',
#    return a positive integer equal to (count('*') - count('#')).
#
# 2. If the number of '#' is greater than '*',
#    return a negative integer equal to (count('*') - count('#')).
#
# 3. If both counts are equal, return 0.
#
# Input:
# ------
# A single string S containing only '*' and '#'.
#
# Output:
# -------
# An integer representing the difference between the number
# of '*' and '#' characters.
#
# Example 1:
# Input: "###***"
# Output: 0
# Explanation: Number of '*' = 3, Number of '#' = 3 → Equal.
#
# Example 2:
# Input: "****##"
# Output: 2
# Explanation: Number of '*' = 4, Number of '#' = 2 → 4 - 2 = 2.
#
# Example 3:
# Input: "##***"
# Output: 1
# Explanation: Number of '*' = 3, Number of '#' = 2 → 3 - 2 = 1.


s=input()
star=0
hash=0
for char in s:
    if(char=="*"):
        star+=1
    else:
        hash+=1
if(hash>star or hash < star):
    a = hash-star
    if(a>0):
        print(f"no of star required:{a}")
    else:
        b=abs(a)
        print(f"no of hash required:{b}")
    
else:
    print("both equal ans 0")
```

```python
# Q2. Problem Statement:
# ----------------------
# Given an integer array Arr of size N, the task is to find the
# count of elements whose value is greater than all of its prior elements.
#
# Note:
# The 1st element of the array should always be considered in the result.
#
# An element Arr[i] is counted if it is strictly greater than
# every element before it (Arr[0] to Arr[i-1]).
#
# Input Format:
# -------------
# 1. An integer N → represents the size of the array.
# 2. Next N lines (or space-separated values) → elements of the array.
#
# Output:
# -------
# Print a single integer representing the count of such elements.
#
# Example 1:
# ----------
# Input:
# N = 5
# Arr = {7, 4, 8, 2, 9}
#
# Explanation:
# 7 → First element, always counted.
# 4 → Not greater than 7 (not counted).
# 8 → Greater than 7 and 4 (counted).
# 2 → Not greater than previous elements (not counted).
# 9 → Greater than all previous elements (counted).
#
# Total elements satisfying the condition = 3
#
# Output:
# 3
#
# Example 2:
# ----------
# Input:
# N = 4
# Arr = {1, 2, 3, 4}
#
# Explanation:
# 1 → counted
# 2 → greater than 1 (counted)
# 3 → greater than 1,2 (counted)
# 4 → greater than 1,2,3 (counted)
#
# Output:
# 4

n=int(input())
arr=list(map(int,input().split()))[:n]
print(arr)



count=1
for i in range(1,n):
    for j in range(i):
        if(arr[i]>arr[j]):
            count+=1
            
  
print(count)        
```

```python
# Gcd for n nummbers peehle 2 number ka gcd nikalo then us gcd ko next number ke sath gcd nikalo and so on.
def gcd(a, b):
    while b != 0:
        temp=b
        b=a%b
        a=temp
    return a


n = int(input())

result = int(input())

for i in range(n - 1):
    num = int(input())
    result = gcd(result, num)

print(result)

# time complexity of gcd is O(log(min(a,b))) and for n numbers it will be O(n*log(max(a,b))) where max(a,b) is the largest number among the n numbers.
# for two numbers it is o(log n)
```

## Q-11 LCM of Two Numbers lcm(12,18) = 36

```python
a=int(input())
b=int(input())

x=a
y=b
# gcd
while y!=0:    # ex gcd(48,18) 48%18=12 , 18%12=6 , 12%6=0
    temp=y  # step 1 temp=18,     step 2 temp=12,   step 3 temp=6
    y=x%y   # step 1 y=48%18=12,  step 2 y=18%12=6, step 3 y=12%6=0
    x=temp #  step 1 x=18,        step 2 x=12,      step 3 x=6
gcd=x
print(f"GCD:{x}")

lcm=(a*b)//(gcd)
print(f"lcm{lcm}")
```

# Q-15 Check Anagram
- Two strings are made using the same characters with the same frequency, but possibly in a different order.


```python
s1=input().lower()
s2=input().lower()
if(len(s1)!=len(s2)):
    print("Not Anagram")
else:
    freq={}
    
    for ch in s1:
        if(ch in freq):
            freq[ch]+=1
        else:
            freq[ch]=1
    for ch in s2:
        if(ch in freq):
            freq[ch]-=1
        else:
            print("Not anagram")
            break
    else:
        anagram=True
        for k in freq: # or use freq.values return only values , or freq.items to directly give k and v
            if(freq[k]!=0):
                print("Not anagram")
                anagram=False
                break
            
        if(anagram):
            print("Anagram") 
```
