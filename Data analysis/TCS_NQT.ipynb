```python
# Demonstrating sorting options
# use .sort() to sort in-place or sorted() to return a new list
a = [3, 1, 4, 2, 5]
print("original:", a)

a.sort()
print("after a.sort():", a)

b = sorted(a, reverse=True)
print("using sorted(reverse=True):", b)
```
