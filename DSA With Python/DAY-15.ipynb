## Rearrange elements by sign

```python
nums=list(map(int,input().split()))
n=len(nums)
li1=[]
li2=[]
li=[]
for n in nums:
    if n<0:
        li2.append(n)
    if n>0:
        li1.append(n)
for i in range(len(li1)):
    li.append(li1[i])
    li.append(li2[i])
print(li)
## TC=o(2n)
## SC=o(2n)
```

```python
## optimal
nums=list(map(int,input().split()))
n=len(nums)
res=[0]*n
p,neg=0,1   
for i in range(n):
    if(nums[i]>=0):
        res[p]=nums[i]
        p+=2
    if(nums[i]<0):
        res[neg]=nums[i]
        neg+=2
print(res)
        
    
```
