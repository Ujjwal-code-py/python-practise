```python
### Print factors of any no
num=int(input())
factors=[]
    
for i in range(1,num+1):
    if((num%i) == 0):
        a=i
        factors.append(a)
print(factors)
    
```

```python
## M-2 iterate to the half of that no

num=int(input())
factors=[]
for i in range(1,int(num/2)+1):
    if(num%i==0):
        factors.append(i)
factors.append(num)
print(factors)
    
```

```python
### optimal solution (Using Root of num)
num=int(input())
factors=[]
for i in range(1, int(num**0.5)+1):
    if(num%i==0):
        
        a=num//i
        factors.append(i)
        if(i!=a):
            factors.append(a)
print(factors)
    
```

## 2427 number-of-common-factors (leet code)

```python
class Solution:
    def commonFactors(self, a: int, b: int) -> int:
        fact=0
        if(a>b):
            num=a
        else:
            num=b
        for i in range(1,int(num//2)+1):
            if(a%i==0 and b%i==0):
                fact+=1
        return fact
a=Solution().commonFactors(a=12,b=6)
print(a)
```
