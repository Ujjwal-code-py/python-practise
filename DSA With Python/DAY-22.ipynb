### Four sum

```python
### Brute Force Approach
nums=[1,0,-1,0,-2,2]
my_set=set()
n=len(nums)

for i in range(n):
    for j in range(i+1,n):
        for k in range(j+1,n):
            for l in range(k+1,n):
                if(nums[i]+nums[j]+nums[k]+nums[l]==0):
                    li=[nums[i],nums[j],nums[k],nums[l]]
                    li.sort()
                    my_set.add(tuple(li))
print([list(ans) for ans in my_set])

## TC=o(n^4)
## SC=o(n)
```

```python
### Better approach
nums=[1,0,-1,0,-2,2]
n=len(nums)
res=set()
for i in range(n):
    
    for j in range(i+1,n):
        my_set=set()
        for k in range(j+1,n):
            fourth = -(nums[i]+nums[j]+nums[k])
            if(fourth in my_set):
                temp=[nums[i],nums[j],nums[k],fourth]
                temp.sort()
                res.add(tuple(temp))
            my_set.add(nums[k])
print([list(ans) for ans in res])
            
## TC=o(n^3)
```

```python
#### optimal Solution
nums=[1,0,-1,0,-2,2]
n=len(nums)
res=[]
target=0
nums.sort()
for i in range(n):
    if(i>0 and nums[i]==nums[i-1]):
        continue
    for j in range(i+1,n):
        if(j>i+1 and nums[j]==nums[j-1]):
            continue
        k=j+1
        l=n-1
        while(k<l):
            total=nums[i]+nums[j]+nums[k]+nums[l]
            if(total==target):
                res.append([nums[i],nums[j],nums[k],nums[l]])
                k+=1
                l-=1
                while(k<l and nums[k]==nums[k-1]):
                    k+=1
                while(k<l and nums[l]==nums[l+1]):
                    l-=1
            elif(total<target):
                k+=1
            else:
                l-=1
print(res)

```
