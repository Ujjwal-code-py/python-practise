```python
## Find the missing number in a list
nums=list(map(int,input().split()))

for i in range(len(nums)+1):
    if(i not in nums):
        print(i)
        break
## Time complexity(o(n^2))
```

```python
## Better
nums=list(map(int,input().split()))
n=len(nums)
freq={}
for i in range(n+1):
    freq[i]=0
for num in nums:
    freq[num]=1
for j in freq :
    if(freq[j]==0):
        print(j)
        break
# TC= o(3n)
```

```python
## optimal find thee sum then subtract
nums=list(map(int,input().split()))
sum_of_list_e=sum(nums)
s_n=0
for n in range(len(nums)+1):  ## sum of n element = n(n+1)/2 you can use this
    s_n+=n
print(s_n-sum_of_list_e)

## TC = o(2n)
```

## Max consecutive 1

```python
nums=list(map(int,input().split()))
max_count=0
count=0
for i in range(len(nums)):
    if(nums[i]==1):
        count+=1
        if(count>max_count):
            max_count=count
        
    else:
        count=0
        
print(max_count)
        
    
```

```python
## optimal
nums=list(map(int,input().split()))
max_count=0
count=0
for i in range(len(nums)):
    if(nums[i]==1):
        count+=1

        
    else:
        max_count=max(count,max_count)
        count=0
        
print(max(max_count,count))

## TC(o(n))
```
