### Longest Consecutive Sequence

```python
## brute force  
nums=list(map(int,input().split()))
n=len(nums)
max_count=0
for i in range(n):
    count=1
    a=nums[i]
    # for j in range(n):
    #     if(a in nums):
    #         count+=1
    #         a+=1
    #     else:
    #         break
    while a+1 in nums:
        count+=1
        a+=1
    if(count>max_count):
        max_count=count       
print(max_count)
## TC: O(n^2) in worst case when all elements are same
```

```python
## better approach 
## using sort and least element of the sequence
nums=list(map(int,input().split())) 
nums.sort() ## o(nlogn)
n=len(nums)
last_smallest=float('-inf')
count=0
longest=0
for i in range(n):
    num=nums[i]
    if((num-1) == last_smallest):
        count+=1
    elif(num!=last_smallest):
        count+=1
        last_smallest=num
    if(count>longest):
        longest=count
print(longest)
## o(nlogn+n)
```

```python
## optimal
nums=list(map(int,input().split()))
n=len(nums)
my_set=set()
for i in range(n):
    my_set.add(nums[i])
longest=0
for num in my_set:
    if(num-1 not in my_set):
        x=num
        count=1
        while(x+1 in my_set):
            count+=1
            x+=1
        longest=max(count,longest)
print(longest)
## o(3n)
```
