## Parameterized & functional recursion

## Parameterized recursion

```python
## sum of 1 to n
def sum_1_to_n(sum,i,N):
    if(i>N):
        print(sum)
        return
    sum_1_to_n(sum+i,i+1,N)
sum_1_to_n(0,0,10)
```

## Functional Recursion
- Create a flow
- Create the base condition

```python
def sum_1_to_n(N):
    if(N==1):
        return 1
    
    return N+sum_1_to_n(N-1)
sum_1_to_n(10)
```

```python
## Factorial of No
def factorial(N):
    if(N==1 or N==0):
        return 1
    
    return (N*factorial(N-1))
factorial(5)
```

```python
## Reverse an array
arr=list(map(int,input().split()))
def reverse_arr(arr,l,r):
    if(l>r):
        print(arr)
        return
    arr[l],arr[r]=arr[r],arr[l]

    reverse_arr(arr,l+1,r-1)
    
r=len(arr)
reverse_arr(arr,2,5)
```

```python
## Check if string is a palindrome or not

def palindrome(word,l,r):
    if(word[l]!=word[r]):
        return False
    if(l>=r):
        return True
    
    return palindrome(word,l+1,r-1)

    

word="wow"
r=len(word)
palindrome(word,0,r-1)
    
```

```python
## fibbo nacci no
def fibbonaci(N):
    if(N<=0):
        return 0
    if(N==1):
        return 1
    return fibbonaci(N-1)+fibbonaci(N-2)
fibbonaci(7)
```
