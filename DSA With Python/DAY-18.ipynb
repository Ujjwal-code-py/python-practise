### Set Matrix Zeros

```python
## Brute force use infinite

nums=[[1,2,3,0],[4,5,0,0],[7,8,9,8]]
r=len(nums)
c=len(nums[0])
for i in range(r):
    for j in range(c):
        if(nums[i][j]==0):
            for k in range(c):
                nums[i][k]=float('inf')
            for l in range(r):
                nums[l][j]=float('inf')
for i in range(r):
    for j in range(c):
        if(nums[i][j]==float('inf')):
            nums[i][j]=0
print(nums)

## Time complexity: O(m*n*(m+n))
```

```python
## Optimal
nums=[[1,2,3,0],[4,5,0,0],[7,8,9,8]]
r=len(nums)
c=len(nums[0])
r_track=[0]*r
c_track=[0]*c
for i in range(r):
    for j in range(c):
        if(nums[i][j]==0):
            r_track[i]=-1
            c_track[j]=-1
for i in range(r):
    for j in range(c):
        if(r_track[i]==-1 or c_track[j]==-1):
            nums[i][j]=0
print(nums)
## TC: O(m*n)
```
