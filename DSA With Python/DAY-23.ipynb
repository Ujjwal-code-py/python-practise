#### Introduction to binary search

```python
## Binary search work for sorted data only
### mid=(high+low)//2
### if mid*mid==x: return mid
nums=[2,4,5,6,8,9,10,12,24]
## binary search
def binarySearch(nums,target):
    n=len(nums)
    low= 0
    high=n-1
    
    while low <= high:
        mid = (low + high) // 2
        if(nums[mid]==target):
            return mid
        elif(nums[mid]<target):
            low=mid+1
        else:
            high=mid-1
        
    return -1

            
            
a=binarySearch(nums,2)
print(a)
        
        
```

```python
## Using recursion
nums=[2,4,5,6,8,9,10,12,24]
n=len(nums)
target=12
def binarySearch(nums,low,high):
    if(low>high):
        return -1
    mid=(low+high)//2
    if(nums[mid]==target):
        return mid
    elif(nums[mid]<target):
        return binarySearch(nums,mid+1,high)
    else:
        return binarySearch(nums,low,mid-1)
  
    

a=binarySearch(nums,0,8)
print(a)
        
## TC=logn
# when n is the no of elements in the list  
        
# Sc=o(1)
```
