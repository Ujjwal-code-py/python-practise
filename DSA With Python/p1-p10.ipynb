```python
######## Check number if the no is palindrome or not ########

N=int(input("Enter a no:"))
num=N
result=0
while(num>0):
    ld=num%10
    result=((result*10)+ld)
    num=num//10
    
if(result==N):
    
    print(result)
    print("palindrome")
else:
    print("Not palindrpme")
    
# TC=log_10(N)
# sc=(o(1))
```

```python
###### Arm Strong number #######
# what is arm strong no  ex no is 153 = 1^3+5^3+3^3 =153 (sum of each digit with the power of its total length)
# 1634=1^4+6^4+3^4+4^4

N=int(input("Enter a no:"))
num=N
n=len(str(N))
result=0
while(num>0):
    ld=num%10
    result=result+(ld**(n))
    num=num//10
    print(result)
if(result==N):
    print("armstrong")
    print(result)

else:
    print("not armstrong")
```

```python
# Prime factor of optimal solution  
from math import sqrt   
n = int(input("Enter no: ")) # 24
res=[]
for i in range(1,int(sqrt(n))+1): # sqrt(24)=4.xx (1,4)
    if((n%i) == 0): # 24%1==0
        res.append(i) # res[1]
        if(n//i!=i): # 24//1 !=1 
            res.append(n//i) # res[1,24] similar flow is 2 and 3 and 4
            
print(res) 

# TC=sqrt(n) and sc=o(1)
# for sorted output res.sort()(TC=o(nlogn))
```

# Part 10

```python
# Frequency Map/ Dictionary
n=list(map(int,input("Enter no: ").split()))
c={}
for i in range(0,len(n)) :
    if n[i] in c:
        c[n[i]]+=1
    else:
        c[n[i]]=1
    
print(c)
```

```python
# another way
# Frequency Map/ Dictionary
n=list(map(int,input("Enter no: ").split()))
c={}
for i in range(0,len(n)):
    c[n[i]]=c.get(n[i],0)+1
print(c)
```
