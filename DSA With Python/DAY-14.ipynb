##  Buy and sell stock

```python
nums=list(map(int,input().split()))
max_profit=0
print(nums)
for i in range(len(nums)):
    profit=nums[i]
    for j in range(i+1,len(nums)):
           if(nums[j]>nums[i]):
            profit=nums[j]-nums[i]
            if(profit>max_profit):
                max_profit=profit
        
print(max_profit)
# TC: O(n^2)
# SC: O(1)
```

```python
## optimal
nums=list(map(int,input().split()))
max_profit=0
min_price=float('inf')
for i in range(len(nums)):
    if(nums[i]<min_price):
        min_price=nums[i]
    if(nums[i]>max_profit):
        max_profit=nums[i]-min_price
print(max_profit)
## TC = o(n)
```
