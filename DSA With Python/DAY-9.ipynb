## Right Rotate the array by single place

```python
## using slicing

nums=[1,2,3,4,5]
a=len(nums)

nums[:]= [nums[a-1]] + nums[0:a-1]
    
print(nums)
```

```python
## using for loop
nums=[1,2,3,4,5]
n=len(nums)
temp=nums[n-1]
for i in range(n-2,-1,-1):
    nums[i+1]=nums[i]
nums[0]=temp
print(nums)
    
```

```python
## Right Rotate on array by k places
nums=list(map(int,input().split()))
n=len(nums)

k=int(input())
for _ in range(k):
    
    temp=nums[n-1]
    for j in range(n-2,-1,-1):
        nums[j+1]=nums[j]
    nums[0]=temp
print(nums)

## TC = (o(k*n))
# sc=o(1)
```

```python
## optimal solution
nums=list(map(int,input().split()))
n=len(nums)

k=int(input())
nums[:]=nums[n-k:]+nums[0:k-1]
print(nums)
```

```python
## optimal solution
def reverse(nums,left,right):
    while left<right:
        nums[left],nums[right]=nums[right],nums[left]
        left+=1
        right-=1
    return nums
def rotate(nums,k):
    n=len(nums)
    reverse(nums,n-k,n-1)
    reverse(nums,0,n-k-1)
    reverse(nums,0,n-1)
    return nums
rotate([1,2,3,45],2)
# TC=o(k/2,(n-k)/2,n/2)=o(n)
# sc=o(1)
```

## Move zeros to end

```python
## Brute force
nums=list(map(int,input().split()))
count=0

for n in nums:
    if(0 in nums):
        count+=1
        nums.remove(0)
for _ in range(count):
    nums.append(0)
print(nums)

```

```python
## Brute force
nums=list(map(int,input().split()))
temp=[]

for i in range(len(nums)):
    if(nums[i]!=0):
        temp.append(nums[i])
for j in range(len(temp)):
    nums[j]=temp[j]
for l in range(len(temp),len(nums)):
    nums[l]=0
print(nums)
    

```

```python
## optimal solution
nums=list(map(int,input().split()))
if(len(nums)==1):
    print(nums)
i=0
while i<len(nums):
    if(nums[i]==0):
        break
    i+=1
if(i==len(nums)):
    print(nums)
j=i+1
while j<len(nums):
    if(nums[j]!=0):
        nums[i],nums[j] = nums[j],nums[i]
        i+=1    
    j+=1
print(nums)

# TC=o(n)
# sc=o(1)
    
```
