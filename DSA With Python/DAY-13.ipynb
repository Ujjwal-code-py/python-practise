### Find the maximum subarray sum

```python
nums=list(map(int,input().split()))
max_sum=float('-inf')
print(nums)
for i in range(0,len(nums)):
    sum=0
    for j in range(i,len(nums)):
        sum=sum+nums[j]
        if(sum>max_sum):
            max_sum=sum

print(max_sum)
# TC: O(n^2)
#SC: O(1)
```

```python
## optimal solution (Kadane's algorithm)
nums=list(map(int,input().split()))
max_sum=float('-inf')
print(nums)
for i in range(0,len(nums)):
    sum=sum+nums[i]
        
    if(sum>max_sum):
        max_sum=sum
    if(sum<0):
        sum=0
print(max_sum)
```
