## Introduction to hashing

```python
## you two list num1 and num2 we have to find that num2 elements are present in num1 at how many times

num1=list(map(int,input().split()))
num2=list(map(int,input().split()))
print(len(num1))
print(len(num2))
count_map={}
for i  in range(len(num2)):
    count=0
    for j in range(len(num1)):
        if(num2[i]==num1[j]):
            count+=1
    count_map[num2[i]]=count
print(count_map)
        


## Time complexity=o(m*N)




```

```python
# M-2 by making a hash list (becuase num1 = 1<=num[i]<=10)
num1=list(map(int,input().split()))
num2=list(map(int,input().split()))
hash_list=[0]*11 #[0,0,--- ,0]
count_map={}
for num in num1:
    hash_list[num]+=1
for num in num2:
    if(num<1 or num>10):
        count_map[num]=0
    else:
        count_map[num]=hash_list[num]
print(count_map)
        
# Time complexity o(m+n)


```

```python
##  Charecter hashing
char=input()
a=list(map(str,input().split()))
count_map={}
for n in a:
    count=0
    for m in char:
        if(n==m):
            count+=1
    count_map[n]=count
print(count_map)
```

```python
## m-2 optimal solution   a<=char[i]<=z  use ascii val 97 to 122
char=input()
a=list(map(str,input().split()))
hash_list=[0]*26
count_map={}
for n in char:
    asc_val=ord(n)
    index=asc_val-97
    hash_list[index]+=1
for m in a:
    asc_val=ord(m)
    index=asc_val-97
    count_map[m]=hash_list[index]
print(count_map)
```
