### Store frequency in dictionary

```python
n=int(input("No of elements: "))
nums=[int(input()) for a in range(1,n+1)] ## input list from user using list comprehension

print(nums)
freq_map={}
for i in range(len(nums)):
    if nums[i] in freq_map:
        freq_map[nums[i]]+=1
    else:
        freq_map[nums[i]]=1
print(freq_map)
    
```

```python
## Count the num comes in the list
n=int(input("No of elements: "))
nums=[int(input()) for a in range(1,n+1)] ## input list from user using list comprehension
count_num=int(input())
print(nums)
freq_map={}
for i in range(len(nums)):
    if nums[i] in freq_map:
        freq_map[nums[i]]+=1
    else:
        freq_map[nums[i]]=1
print(freq_map)
if(count_num not in freq_map):
    print("error")
else:
    print(f"no of count {count_num}:{freq_map[count_num]}")
    
```
