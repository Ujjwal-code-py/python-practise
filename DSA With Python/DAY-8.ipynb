## Remove duplicates from sorted array (Leetcode 26)
- put all unique elements in sorted array in start inplace
- return total count of unique element

```python

## Brute force


def remove_duplicates(arr):
    freq_map={}
    for i in range(len(arr)):
        freq_map[arr[i]]=0
    count=0
    for j in freq_map:
        arr[count]=j
        count+=1
    return count

remove_duplicates([1,2,3,4,5,5,6,6])
```

```python
## optimal
def remove_duplicates(arr):
    if(len(arr)==1):
        return 1
    i=0
    j=i+1
    while(j<len(arr)):
        if(arr[j]!=arr[i]):
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
        j+=1
            
            
    return(i+1)
remove_duplicates([1,2,3,4,5,5,6,6])
```
