```python
# Introduction to hashing
# check the elements in m matching in n how many times 
# 1<=n[i]<=10
n=list(map(int,input().split()))
m=list(map(int,input().split()))
print(n)
print(m)
hash={}
for i in m:
    count=0
    for k in n:
        if(i==k):
            count+=1
    hash[i]=count
print(hash)
```

```python
# optimal solution
hash_list=[0]*(len(n)+1)
n=list(map(int,input().split()))
m=list(map(int,input().split()))
print(n)
print(m) # m<n
hash={}
for i in n:
    hash_list[i]+=1
for k in m:
    if k<1 or k>10: 
        print(0)
    else:
        hash[k]=hash_list[k]
print(hash)
```

```python
# Frquency dictionary method

n=list(map(int,input().split()))
m=list(map(int,input().split()))
print(n)
print(m) 
dict={}
hash={}
for i in n:
    dict[i]=dict.get(i,0)+1
for k in m:
    if k in dict:
        hash[k]=dict[k]
print(hash)
print(dict)
```

```python
## Character Hashing
s=input()
n=list(map(str,(input().split())))
print(s)
print(n)
hash={}
for a in n:
    count=0
    for k in s:
        if a==k:
            count+=1
    hash[a]=count
print(hash)
    
    
```

```python
# Introduction to recursion
# in recursion when you calls a function which runs infinite. in pyhton it run arount 987 times until the stack limit is reached.
# in otheer languages like c, c++ it gives stack overflow error
```

```python
# Head Rescursion, Tail recusion.
# in head recursion the recursive call is the first statement in the function and in tail recursion the recursive call is the last statement in the function.
def head_recursion(n):
    if n>0:
        print(n)
        head_recursion(n-1)
def tail_recursion(n):
    if n>0:
        tail_recursion(n-1)
        print(n)
        
head_recursion(5)
tail_recursion(5)
```
