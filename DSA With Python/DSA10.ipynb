```python
## Brute force
## Search the target element in the list and return the index

nums=list(map(int,input().split()))
target=int(input())

for i in range(len(nums)):
    if(nums[i]==target):
        print(i)
        break
 
```

## Merged two sorted arrays but it should not contain unique element

```python
num1=list(map(int,input().split()))
num2=list(map(int,input().split()))
res=[]
i,j=0,0
n,m=len(num1),len(num2)

while(i<n and j<m):
    if(num1[i]<=num2[j]):
        if(len(res)==0 or res[-1]!=num1[i]):
            res.append(num1[i])
        i+=1
    else:
        if(len(res)==0 or res[-1]!=num2[j]):
            res.append(num2[j])
        
        j+=1
while(i<n):
        if(len(res)==0 or res[-1]!=num1[i]):
            res.append(num1[i])
        i+=1
while(j<m):
        if(len(res)==0 or res[-1]!=num2[j]):
            res.append(num2[j])
        j+=1
print(res)

## TC = o(m+n)
## SC =O(1)
```
