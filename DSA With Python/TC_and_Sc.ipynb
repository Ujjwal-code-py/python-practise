# TC of different operations

```python
# ===============================
# Python List Operations – Time Complexity
# ===============================
#
# Operation               | Average Case | Amortized Worst Case
# -------------------------------------------------------------
# Copy                    | O(n)         | O(n)
# Append                  | O(1)         | O(1)
# Pop last                | O(1)         | O(1)
# Pop intermediate        | O(n)         | O(n)
# Insert                  | O(n)         | O(n)
# Get Item (arr[i])       | O(1)         | O(1)
# Set Item (arr[i] = x)   | O(1)         | O(1)
# Delete Item             | O(n)         | O(n)
# Iteration               | O(n)         | O(n)
# Get Slice               | O(k)         | O(k)
# Delete Slice            | O(n)         | O(n)
# Set Slice               | O(k + n)     | O(k + n)
# Extend                  | O(k)         | O(k)
# Sort                    | O(n log n)   | O(n log n)
# Multiply                | O(n * k)     | O(n * k)
# x in s                  | O(n)         | O(n)
# min(s), max(s)          | O(n)         | O(n)
# Get Length (len(s))     | O(1)         | O(1)
#
# Notes:
# n = number of elements in the list
# k = number of elements being added/sliced
# Amortized means average cost over many operations
```

```python
# ===============================
# Python Dictionary Operations – Time Complexity
# ===============================
#
# Operation               | Average Case | Amortized Worst Case
# ------------------------------------------------------------
# k in d                  | O(1)         | O(n)
# Copy                    | O(n)         | O(n)
# Get Item (d[k])         | O(1)         | O(n)
# Set Item (d[k] = v)     | O(1)         | O(n)
# Delete Item (del d[k])  | O(1)         | O(n)
# Iteration               | O(n)         | O(n)
#
# Notes:
# n = number of key-value pairs in the dictionary
# Worst case happens due to hash collisions
```

```python
# ===============================
# Python Set Operations – Time Complexity
# ===============================
#
# Operation                                   | Average Case                     | Worst Case
# -----------------------------------------------------------------------------------------
# x in s                                      | O(1)                             | O(n)
# Union (s | t)                               | O(len(s) + len(t))               | O(len(s) + len(t))
# Intersection (s & t)                        | O(min(len(s), len(t)))           | O(len(s) * len(t))
# Multiple Intersection (s1 & s2 & ... & sn)  | O((n-1) * l)                     | O((n-1) * l)
# Difference (s - t)                          | O(len(s))                        | O(len(s))
# s.difference_update(t)                     | O(len(t))                        | O(len(t))
# Symmetric Difference (s ^ t)                | O(len(s))                        | O(len(s) * len(t))
# s.symmetric_difference_update(t)           | O(len(t))                        | O(len(t) * len(s))
#
# Notes:
# len(s), len(t) = sizes of sets
# l = max(len(s1), len(s2), ..., len(sn))
# Worst case happens due to hash collisions
```
