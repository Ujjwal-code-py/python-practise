## 3 SUM
##### nums[i]+nums[j]+nums[k]=0
##### i!=j!=k

```python
nums=list(map(int,input().split()))
my_set=set()
n=len(nums)

for i in range(n):
    for j in range(i+1,n):
        for k in range(j+1,n):
            if(nums[i]+nums[j]+nums[k]==0):
                li=[nums[i],nums[j],nums[k]]
                li.sort()
                my_set.add(tuple(li))
print([list(ans) for ans in my_set])

#### TC=o(n^3)
```

```python
## Better Approach
# nums[k]= -(nums[i]+nums[j])
```

```python

arr=list(map(int,input().split()))
n=len(arr)
res=set()
for i in range(n):
    my_set=set()

    for j in range(i+1,n):
        third= -(arr[i]+arr[j])
        if(third in my_set):
            temp=[arr[i],arr[j],third]
            temp.sort()
            res.add(tuple(temp))
        my_set.add(arr[j])
print([list(ans) for ans in res])
## TC=o(n^2)
## sc=o(n)+o(no of triplet)

```

```python
# optimal solution 
arr=list(map(int,input().split()))
n=len(arr)
ans=[]
arr.sort()

for i in range(n):
    if(i!=0 and arr[i]==arr[i-1]):
        continue
    j=i+1
    k=n-1
    while j<k:
        total_sum=arr[i]+arr[j]+arr[k]
        if(total_sum<0):
            j+=1
        elif(total_sum>0):
            k-=1
        else:
            temp=[arr[i],arr[j],arr[k]]
            ans.append(temp)
            j+=1
        # skip duplicates
        while j<k and nums[j]==nums[j-1]:
            j+=1
        
        while j<k and nums[k]==nums[k+1]:
            k-=1
print(ans)         

# TC=o(n^2)
```
