## Selection Sort

```python
# sort the array in ascending order
nums=list(map(int,input().split()))
def selection_sort(nums):
    n=len(nums)
    for i in range(n):
        min_indx=i
        for j in range(i+1,n):
            if(nums[j]<nums[min_indx]):
                min_indx=j
        nums[i],nums[min_indx]=nums[min_indx],nums[i]
    return print(nums)
selection_sort(nums)
```

```python
## Sort the array in descendign order
nums=list(map(int,input().split()))
def selection_sort(nums):
    n=len(nums)
    for i in range(n):
        max_indx=i
        for j in range(i+1,n):
            if(nums[j]>nums[max_indx]):
                max_indx=j
        nums[i],nums[max_indx]=nums[max_indx],nums[i]
    return print(nums)
selection_sort(nums)
```

## Bubble sort (adjacent swap)

```python
## sort in ascending order
nums=list(map(int,input().split()))
def bubble_sort(nums):
    n=len(nums)
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if(nums[j]>nums[j+1]):
                nums[j],nums[j+1]=nums[j+1],nums[j]
    return print(nums)
bubble_sort(nums)
```

```python
## sort in descending order
nums=list(map(int,input().split()))
def bubble_sort(nums):
    n=len(nums)
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if(nums[j]<nums[j+1]):
                nums[j],nums[j+1]=nums[j+1],nums[j]
    return print(nums)
bubble_sort(nums)

```

## Insertion Sort

```python
## sort in ascending order
nums=list(map(int,input().split()))
def insertion_sort(nums):
    n=len(nums)
    for i in range(1,n):
        key=nums[i]
        j=i-1
        while(j>=0 and nums[j]>key):
            nums[j+1]=nums[j]
            j-=1
        
        nums[j+1]=key
    return print(nums)
insertion_sort(nums)
```

```python
## sort in descnding order
nums=list(map(int,input().split()))
def insertion_sort_desc(nums):
    n=len(nums)
    for i in range(1,n):
        key=nums[i]
        j=i-1
        while(j>=0 and nums[j]<key):
            nums[j+1]=nums[j]
            j-=1
            
        nums[j+1]=key
    return print(nums)
insertion_sort_desc(nums)
```
